## Missing & Wrong Datas

```python
print("Total Null:", df.isna().sum().sum())
df.isna().sum()
```

```python
df[df['Agriculture'].isnull()]['Country']
```

```python
df[df["Agriculture"].isna()] = df[df["Agriculture"].isna()].fillna(0)
```

```python
print("Toplam Boş:", df.isna().sum().sum())
df.isna().sum()
```

```python
df[df["Literacy (%)"].isnull()]
```

```python
df.groupby("Region")["Literacy (%)"].mean().sort_values(ascending=False)
```

```python
df["Literacy (%)"] = df["Literacy (%)"].fillna(df.groupby("Region")["Literacy (%)"].transform("mean"))
```

```python
df[df["Literacy (%)"].isnull()]
```

```python
df.iloc[99]["Literacy (%)"]
```

```python
print("Toplam Boş:", df.isna().sum().sum())
df.isna().sum()
```

```python
df["Climate"] = df["Climate"].fillna(df.groupby("Region")["Climate"].transform("mean"))
```

```python
print("Total Null:", df.isna().sum().sum())
df.isna().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
# Calculate the mean for each region only for the numeric columns
region_means = df.groupby('Region')[numeric_cols].transform('mean')

# Fill the NA values in the original df with the means from the same region
df[numeric_cols] = df[numeric_cols].fillna(region_means)
```

```python
print("Toplam Boş:", df.isna().sum().sum())
df.isna().sum()
```

## Model Preparation

```python
X = df.drop("Country", axis=1)
```

```python
X = pd.get_dummies(X, drop_first=True)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaled_X = StandardScaler().fit_transform(X)
scaled_X
```

```python
pd.DataFrame(scaled_X)
```

---
## K - Means Model

```python
from sklearn.cluster import KMeans
```

```python
ssd = []

for k in range(2,50):
    model = KMeans(n_clusters=k)
    model.fit(scaled_X)

    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,50), ssd, "o")
plt.xlabel("K")
plt.ylabel("Noktalar Arası Mesafelerin\n Karelerinin Toplamı")
```

```python
ssd_seri = pd.Series(ssd)
```

```python
plt.plot(range(2,50), ssd_seri.diff())
```

```python
model = KMeans(n_clusters=3)
model.fit(scaled_X)
```

```python
model.labels_
```

```python
X['4 Clusters'] = model.labels_
```

```python
X.corr()['4 Clusters'].sort_values()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(X.corr()["4 Clusters"].sort_values());
plt.xticks(rotation=90)
```
