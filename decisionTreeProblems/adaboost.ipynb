```python
X = df.drop("class", axis = 1)
y = df["class"]
```

```python
X = pd.get_dummies(X,drop_first=True)
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
 test_size=0.1, random_state=9)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier(n_estimators=18)
model.fit(X_train, y_train)
## Dummy Variable olmazsa burada hata verir
```

```python
mushroom_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, accuracy_score
```

```python
ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
```

```python
print(classification_report(y_test, mushroom_pred))
```

```python
toleranceRatio = []

for mushroom_est in range(1,30):
    model = AdaBoostClassifier(n_estimators=mushroom_est)
    model.fit(X_train, y_train)
    mushroom_try_pred = model.predict(X_test)
    
    tolerance = 1 - accuracy_score(y_test, mushroom_try_pred)
    toleranceRatio.append(tolerance)
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,30), toleranceRatio);
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,30), toleranceRatio, lw=5)
plt.xlim(10,20)
plt.ylim(-0.01,0.04)
```
